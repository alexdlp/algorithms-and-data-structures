## Two Sum

Given an array of integers `nums` and an integer `target`, return the **indices of the
two numbers** such that they add up to `target`.

You may assume that each input would have **exactly one solution**, and you may **not
use the same element twice**. You can return the answer in any order.

### Examples

**Example 1**

```
Input:  nums = [2, 7, 11, 15], target = 9
Output: [0, 1]
Explanation: nums[0] + nums[1] == 9, so we return [0, 1].
```

**Example 2**

```
Input:  nums = [3, 2, 4], target = 6
Output: [1, 2]
```

**Example 3**

```
Input:  nums = [3, 3], target = 6
Output: [0, 1]
```


Esta Puede ser la implementación navive. Dos bucles, cada uno O(N), los recorres enteros. Complejidad temporal O(N^2), y espacial O(1).

````python
def two_sum_naive(array: list, target: int) -> list:
    """
    Find two numbers in the array that add up to the target value.
    
    Parameters:
    array (np.ndarray): A numpy array of integers.
    target (int): The target sum value.
    
    Returns:
    list: A list containing the indices of the two numbers that add up to the target value.
          If no such pair exists, returns an empty list.
    """
    result = []
    for i in range(len(array)): ## O(N)
        for j in range(len(array)-1): ## O(N)
            if array[i] + array[j+1] == target:
                result.append(i)
                result.append(j+1)

                return result
            
    return []


print(two_sum_naive([2, 7, 11, 15], target = 9))
print(two_sum_naive([3, 2, 4], target = 6))
print(two_sum_naive([3, 3], target = 6))

assert two_sum_naive([2, 7, 11, 15], target = 9) == [0,1]
assert two_sum_naive([3, 2, 4], target = 6) == [1,2]
assert two_sum_naive([3, 3], target = 6) == [0,1]
````

Esto parece que podría ser otra opción, pero en realidad es lo mismo.

En cada caso, lo que hacemos es calcular cual es el número complementario que necesitamos para llegar al target, y lo buscamos en el array.
De todos modos `array.index(complement)` internamente hace el mismo bucle O(N), por lo que nos quedamos igual.


````python
def two_sum(array: list, target: int) -> list:
    """
    Find two numbers in the array that add up to the target value.
    
    Parameters:
    array (np.ndarray): A numpy array of integers.
    target (int): The target sum value.
    
    Returns:
    list: A list containing the indices of the two numbers that add up to the target value.
          If no such pair exists, returns an empty list.
    """
    for i in range(len(array)): ## O(N)
        complement = target - array[i]
        #print(array[i+1:])
        try:
            index = array.index(complement, i + 1)
            return [i, index]
        except ValueError as ex:
            pass
    
    return []

     
print(two_sum([2, 7, 11, 15], target = 9))
print(two_sum([3, 2, 4], target = 6))
print(two_sum([3, 3], target = 6))

assert two_sum([2, 7, 11, 15], target = 9) == [0,1]
assert two_sum([3, 2, 4], target = 6) == [1,2]
assert two_sum([3, 3], target = 6) == [0,1]
````

Esta si lo mejora.

Lo que hacemos es iterar sobre el array de entrada (O(N)), e ir creando un diccionario con los valores que hemos visto como keys y los índices como values. De manera que podemos buscar los complementos en ese diccionario de vistos, y eso es O(1).

````python
def two_sum_dict(array: list, target: int) -> list:
    """
    Find two numbers in the array that add up to the target value.
    
    Parameters:
    array (np.ndarray): A numpy array of integers.
    target (int): The target sum value.
    
    Returns:
    list: A list containing the indices of the two numbers that add up to the target value.
          If no such pair exists, returns an empty list.
    """
    visto = {}
    for idx, number in enumerate(array): # O(N)

        #print('iteration', i)
        
        complement = target - number
        
        #print(complement)
    
        # check if we have already seen the complement O(1)
        if complement in visto:

            complement_idx = visto.get(complement)
            result = [complement_idx,idx]

            return result
        
        # if we dont, we add it to the hash table
        else:
            visto[number]=idx
        
        #print(visto)

    return []

print(two_sum_dict([2, 7, 11, 15], target = 9))
print(two_sum_dict([3, 2, 4], target = 6))
print(two_sum_dict([3, 3], target = 6))

assert two_sum_dict([2, 7, 11, 15], target = 9) == [0,1]
assert two_sum_dict([3, 2, 4], target = 6) == [1,2]
assert two_sum_dict([3, 3], target = 6) == [0,1]
````

## Move Zeroes

Given an integer array `nums`, move all `0`'s to the end of it while maintaining the
relative order of the non-zero elements.

**Note:** you must do this **in-place**, without making a copy of the array.

### Examples

**Example 1**

```
Input:  nums = [0, 1, 0, 3, 12]
Output: [1, 3, 12, 0, 0]
```

**Example 2**

```
Input:  nums = [0]
Output: [0]
```

Version naive, en la que filtramos los números que no son 0 del array y despues añadimos los ceros que faltan al final. Esta soluciónm sería incorrecta, porque crea un array nuevo, cosa que el ejercicio no nos permite. Complejidad temporal O(N) y espacial O(N) también. 

````python
def move_zeros_naive(array: list):
    """
    Move all zeros in the array to the end while maintaining the order of non-zero elements.

    Parameters:
    array (list): A list of integers.
    Returns:
    list: A new list with all zeros moved to the end.
    """
    original_length = len(array)
    ## filtrado del array O(N)
    array = [num for num in array if num!=0]

    for i in range(original_length-len(array)):
        array.append(0)

    return array


print(move_zeros_naive([0, 1, 0, 3, 12]))
print(move_zeros_naive([0]))

assert move_zeros_naive([0, 1, 0, 3, 12]) == [1, 3, 12, 0, 0]
assert move_zeros_naive([0]) == [0]
````

En esta versión, reescribimos el array in place, siendo por tanto complejidad temporal O(N) debido el bucle de iteración, pero solo O(1) en memoria, porque solo necesitamos el contador de ceros. 

````python
def move_zeros(array: list):
    """
    Move all zeros to the end in place while preserving non-zero order.

    Args:
        array: List of integers.

    Returns:
        The same list, modified in place
    """

    zero_counter = 0
  
    for number in array:

        if number != 0:
            array[zero_counter] = number
            zero_counter +=1
     
    for i in range(zero_counter, len(array)):
        array[i]=0

    return array


print(move_zeros([0, 1, 0, 3, 12]))
print(move_zeros([0]))

assert move_zeros([0, 1, 0, 3, 12]) == [1, 3, 12, 0, 0]
assert move_zeros([0]) == [0]
````
